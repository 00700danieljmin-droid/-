```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from sklearn.linear_model import LinearRegression

# 1. 한글 폰트 설정 및 환경 준비
!sudo apt-get install -qq -y fonts-nanum
!fc-cache -fv
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font_prop = fm.FontProperties(fname=font_path)
plt.rc('font', family='NanumBarunGothic')

try:
    # 2. 데이터 불러오기 (엑셀로 만든 100개 데이터셋)
    df = pd.read_csv('data.csv')

    # AI 학습: 무게(X)를 넣으면 두께(y)가 나오도록 학습
    X = df.iloc[:, [0]].values # 무게 열
    y_thick = df.iloc[:, 5].values # 균사체 두께 열

    model_thick = LinearRegression().fit(X, y_thick)

    print("="*60)
    print(" AI 시스템 가동: KS 표준 데이터 기반 학습이 완료되었습니다.")
    print("="*60)

    # 3. 무게 입력 및 자동 결과 출력
    user_input = float(input("\n측정할 화물의 무게(kg)를 입력하세요: "))

    # 입력된 무게에 맞는 '낙하 높이' 찾기 (데이터셋 내 가장 유사한 값 기준)
    idx = (df.iloc[:, 0] - user_input).abs().idxmin()
    target_height = df.iloc[idx, 1] # 표준 낙하 높이(H) 자동 매칭

    # AI 예측값 산출
    pred_thick = model_thick.predict([[user_input]])[0]
    pred_layer = int(np.ceil(pred_thick / 0.4)) # 뽁뽁이 겹수는 정수 올림

    # 결과 대화창 출력
    print(f"\n[분석 결과]")
    print(f"▶ 입력 무게: {user_input} kg")
    print(f"▶ 표준 낙하 높이(KS 기준): {target_height} m (자동 매칭)")
    print("-" * 30)
    print(f"▶ AI 추천 균사체 설계 두께: {pred_thick:.2f} cm")
    print(f"▶ AI 추천 에어캡(뽁뽁이) 겹수: {pred_layer} 겹 (정수 산출)")
    print("-" * 30)
    print("※ 위 수치는 파손 임계 가속도 50G 미만을 유지하기 위한 최적 설계값입니다.")

except Exception as e:
    print(f"오류: {e}\n'data.csv' 파일이 업로드되어 있는지, 열 순서가 맞는지 확인해주세요.")
```
